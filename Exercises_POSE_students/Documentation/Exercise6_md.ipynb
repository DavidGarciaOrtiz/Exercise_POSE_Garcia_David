**a) Find graphically the RPY Orientation**

We have a quaternion of $ [0, 0.866, 0, 0.5] $ which we can turn into an angle vector by identifying the angle $\theta$ which defines the rotation of the vector to compute the RPY orientation. Following the definition of quaternion, we can compute easily this angle:

$$
q = s + v = s + <v_x, v_y, v_z> = 0 + <0.866, 0, 0.5>
$$
$$
cos(\theta/2) = s = 0
$$
$$
\theta = 2·arccos(s) = 180 ^\circ
$$

Now we can operate like in previous case (exercise 5) by treating the quaternion as an angle vector. So for the vector $ (0.866, 0, 0.5) $ rotated 180 degrees we have the following RPY angles: 

$$
\text{RPY} = [180^\circ, -60^\circ, 0^\circ]
$$

**b) Find Analytically the Rotation matrix representing this orientation.**

The RPY rotation matrix for this orientation is given by the multiplication of **Cardano Angles** following the zyx order:

$$
\text{R} = R_z(Yaw) · R_y(Pitch) · R_x(Roll) =  R_z(60^\circ) · R_y(-90^\circ) · R_x(180^\circ)
$$

Each axis rotation around cardano angles has its associated matrix, so the resulting transformation arises from the product of these matrices:

$$
R = 
\begin{bmatrix}
cos(60^\circ) & -sin(60^\circ) & 0 \\
sin(60^\circ) & cos(60^\circ) & 0 \\
0 & 0 & 1
\end{bmatrix}
· 
\begin{bmatrix}
cos(90^\circ) & 0 & -sin(90^\circ) \\
0 & 1 & 0 \\
-sin(90^\circ) & 0 & cos(90^\circ) 
\end{bmatrix}
· 
\begin{bmatrix}
1 & 0 & 0 \\
0 & cos(180^\circ) & -sin(180^\circ) \\
0 & sin(180^\circ) & cos(180^\circ)
\end{bmatrix}

$$

The rotation matrix describing this orientation is the following:
$$
R = \begin{bmatrix}
0 & 0 & 1 \\
0 & -1 & 0 \\
1 & 0 & 0
\end{bmatrix}

$$

**c) Find the Euler angles and AO-vectors Graphically**
Based on the image of the exercise, we can deduce graphically that the Euler angles in the zyz order are the following:

$$
\text{zyz} = [-180^\circ, 135^\circ,90^\circ]
$$

The AO vectors can be extracted from the rotation matrix, which are in form of column vectors:

$$
R = \begin{bmatrix}
0 & 0 & 1 \\
0 & -1 & 0 \\
1 & 0 & 0
\end{bmatrix}
=
\begin{bmatrix}
n_x & o_x & a_x \\
n_y & o_y & a_y \\
n_z & 0_z & a_z
\end{bmatrix}

$$

Z-axis vector (Approach): 
$$ \hat{a} = (a_x, a_y, a_z) =  (1, 0, 0) 
$$
Y-axis vector (Ortogonal): 
$$ \hat{o} = (o_x, o_y, o_z) = (0, -1, 0) 
$$
X-axis vector: 
$$ \hat{n} = \hat{o} \times \hat{a} = (n_x, n_y, n_z) = (0, 0, 1) $$
