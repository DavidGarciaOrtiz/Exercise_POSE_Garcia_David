**a) Which is the RPY Orientation of “Target B” with respect “UR5e Base”?**

The RPY orientation of **Target B** with respect to the UR5e Base is the following

$$
\text{RPY} = [180^\circ, -45^\circ, 0^\circ]
$$

We observe that the initial brick frame is rotated 180 degrees in x axis with respect the UR5e base frame, so the Target B frame must include this rotation in addition to the -45 degree rotation to be properly placed onto the slope. 

---

---

**b) Find Analytically the Rotation matrix representing this orientation.**

The RPY rotation matrix for this orientation is given by the multiplication of **Cardano Angles** following the zyx order:

$$
\text{R} = R_z(Yaw) · R_y(Pitch) · R_x(Roll) =  R_z(0^\circ) · R_y(45^\circ) · R_x(180^\circ)
$$

Each axis rotation around cardano angles has its associated matrix, so the resulting transformation arises from the product of these matrices:

$$
zyz = 
\begin{bmatrix}
cos(0^\circ) & -sin(0^\circ) & 0 \\
sin(0^\circ) & cos(0^\circ) & 0 \\
0 & 0 & 1
\end{bmatrix}
· 
\begin{bmatrix}
cos(-45^\circ) & 0 & -sin(-45^\circ) \\
0 & 1 & 0 \\
-sin(-45^\circ) & 0 & cos(-45^\circ) 
\end{bmatrix}
· 
\begin{bmatrix}
1 & 0 & 0 \\
0 & cos(180^\circ) & -sin(180^\circ) \\
0 & sin(180^\circ) & cos(180^\circ)
\end{bmatrix}

$$

The rotation matrix describing this orientation is the following:
$$
R = \begin{bmatrix}
\frac{\sqrt{2}}{2} & 0 & \frac{\sqrt{2}}{2} \\
0 & -1 & 0 \\
\frac{\sqrt{2}}{2} & 0 & -\frac{\sqrt{2}}{2}
\end{bmatrix}

$$

**c) Find the Euler angles and AO-vectors Graphically**
Based on the image of the exercise, we can deduce graphically that the Euler angles in the zyz order are the following:

$$
\text{zyz} = [180^\circ, -135^\circ,0^\circ]
$$

The AO vectors can be extracted from the rotation matrix, which are in form of column vectors:

$$
R = \begin{bmatrix}
\frac{\sqrt{2}}{2} & 0 & \frac{\sqrt{2}}{2} \\
0 & -1 & 0 \\
\frac{\sqrt{2}}{2} & 0 & -\frac{\sqrt{2}}{2}
\end{bmatrix}
=
\begin{bmatrix}
n_x & o_x & a_x \\
n_y & o_y & a_y \\
n_z & 0_z & a_z
\end{bmatrix}

$$

Z-axis vector (Approach): $$ \hat{a} = (a_x, a_y, a_z) =  (\frac{\sqrt{2}}{2}, 0, -\frac{\sqrt{2}}{2}) $$
Y-axis vector (Ortogonal): $$ \hat{o} = (o_x, o_y, o_z) = (0, -1, 0) $$
X-axis vector: $$ \hat{n} = \hat{o} \times \hat{a} = (n_x, n_y, n_z) = (\frac{\sqrt{2}}{2}, 0, \frac{\sqrt{2}}{2}) $$
